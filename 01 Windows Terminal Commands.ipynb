


## Windows Terminal Tips

### Introduction

This quick-tip post covers essential CMD commands for Windows OS users. Terminal commands are quite versatile, ranging from managing files, troubleshooting network issues, to optimizing your system. I've picked the most handy commands that help you perform tasks efficiently.

### Opening Command Prompt

You must first open the `Command Prompt` to start using it. Here are some ways to do that:

- (a). `Right-click` the windows start menu icon on the bottom taskbar and select `Run` on the menu that appears.

    

    Type `cmd` in the **Run dialogue box** and press `Enter` as shown below:

    


- (b). Press `Win + R` to open the Run dialogue box.
    Type `cmd` in the **Run dialogue box** and press `Enter`.
    
### Most used terminal commands

- `mkdir hello` - creates a new directory **hello**.

- `cd hello` - changes directory to **hello**, as created above.

- `code .` - opens the current directory as workspace in **vs code**.

- `cd ~` - goes back to home directory.

- `cd ..` - shifts one directory back.

- `rmdir name` - removes directory called **name**.

- `del fname` - deletes a file named **fname** in current directory.

### Rename a folder oldfolder to newfolder

`ren oldfolder newfolder`

### List all files and folders in current directory
`dir` 

### Open directory in windows terminal

**Note:** To open a directory in windows terminal, go to the folder using the usual File Explorer. Select the path in the address bar and type `cmd`, then press enter.




### Open directory as `vs code` workspace

To open the current directory as `vs code` workspace, type the following on windows terminal:
`code .`
