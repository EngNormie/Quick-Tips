


## Anaconda Terminal

### Introduction

Anaconda is a distribution of Python designed specifically for machine learning and data science. There are several pre-installed packages that come with it, such as SciPy, Malplotlib, Pandas, and NumPy. Anaconda is also designed to work seamlessly with Jupyter Notebooks, an open-course web application that enables one to share and create documents that consist of equations, narrative text, visualization, and live code.  Windows uses the Anaconda Prompt CLI to process text commands to do various tasks.

### Pre-requisites
You must have Anaconda installed on your system before you can start using Anaconda Terminal. 

Check for the installation by typing the following command on `windows terminal` or `Powershell`:
    `conda --version`

If no installation is found, visit the [Anaconda Distribution page](https://www.anaconda.com/distribution/) and download the installer for your Operating System.

Once the installer is downloaded, double-click on it and follow the installation instructions. Make sure to check the box that says `"Add Anaconda to my PATH environment variable"` during installation. This step ensures that you can access Anaconda directly from the command prompt.

### Anaconda terminal commands

The following are regular anaconda terminal commands you would encounter in your data science and analytics workflows:


#### Create Environment without specifying packages

Type the following command in Anaconda Terminal:

`conda create -n ENV_NAME`

Replace `ENV_NAME` with the name for your environment


#### Create Environment with required packages

Use the create commanda, accompanied by a list of the preferred modules and appropriate versions.

`conda create -n penguins python=3.11 pandas seaborn matpotlib statsmodels jupyter` 

The above command creates an environment named penguins.

_You can substitute with your preferred environment name. Put correct packages required for your project._


#### Activate Environment 

Once the environment is created, activate it in the terminal using the following command:

`conda activate penguins`	

#### Install packages

`conda install packagename`

Or:

`python -m pip install packagename`

_Replace "packagename" with the actual package you want to install_

#### Switching between environments

When youre ready to switch between projects, simply activate the environment of your other project. Activating a different environment will deactivate your current one.

1.	(Optional) View a list of all your environments by running the following command:

`conda info --envs`

2.	To switch to a different environment, activate it by running the following command:

`conda activate ENV-NAME`

_Replace "ENV-NAME" with the name of the environment you want to switch to._

#### Deactivating an environment

It is best practice to deactivate your environment when you are finished working in it.
To deactivate your active environment, run the following command:

`conda deactivate`

### Summary

Anaconda Prompt CLI allows you to communicate with a computer or operating system using plain text, without interacting with windows, icons or menus. You can perform a wide variety of tasks, such as file management, system operations, program control, etc. Anaconda Prompt lets you perform all Anaconda functions.
